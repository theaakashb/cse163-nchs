# Classwork : The Admissions Black Box

Below is a 30-minute activity called **"The Admissions Black Box."** It uses the lesson's Circle/Square analogy to show how bias enters a system and how hard it is to "fix" it without losing accuracy.

## 1. Setup (5 Minutes)

Divide the class into small groups (3–4 students). Each group is an **Admissions Committee** for "Polygon University."

* **The Data:** Give each group a stack of 10 "Applicant Cards."
* **Circles:** Mostly have high SAT scores (1400+) and 3+ extracurriculars.
* **Squares:** Mostly have mid-range SAT scores (1100–1200) and 1–2 extracurriculars.


* **The Secret "Ground Truth":** Tell the students that on this planet, Squares have no access to SAT tutors or expensive sports, but they are equally "capable" of succeeding in college.

## 2. Phase 1: Human Bias (10 Minutes)

Ask the groups to quickly select 4 students to admit based **only** on the SAT score and extracurriculars.

* **The Result:** Most groups will admit 4 Circles.
* **Discussion:** Ask: "Is this fair? You picked the 'best' scores, but did you pick the 'best' potential?"

## 3. Phase 2: The "Blind" Algorithm (5 Minutes)

Tell them they are now training an AI. They must create a simple "Rule" (an algorithm) for admission.

* **Constraint:** They are **not allowed** to use the "Shape" (Race) variable.
* **Common Rule:** *"Admit if SAT > 1300."*
* **The Reveal:** Show them that because of the zip codes or school names on the cards (which correlate with the shapes), their "blind" rule still only admits Circles.
* **Lesson:** Removing the label doesn't remove the bias.

## 4. Phase 3: The Fairness Trade-off (10 Minutes)

Now, tell the groups they must comply with the **"Fairness Act of 2026."** They must ensure at least 1 Square is admitted.

* **The Conflict:** To admit a Square, they have to reject a Circle with a higher SAT score.
* **The "Pareto" Moment:** Ask the groups:
* "Does your university's 'average SAT' go down? (Accuracy Loss)"
* "Is it 'fair' to the Circle you rejected? (Individual Fairness)"
* "Is it 'fair' to the Square community to ignore their lack of tutors? (Group Fairness)"

---

## **Wrap-up Discussion Questions**

* **The "Why":** If we know the SAT is a biased measure for Squares, should we keep using it?
* **Generative AI:** If an AI wrote the "Acceptance Letters," and it used formal, flowery language for Circles but "simplified" language for Squares because of their data profile, would that be a problem? (This connects to the **Representational Harm** update in your lesson).
* **The Future:** How would a "Human-in-the-loop" (a real person checking the AI's math) have changed the outcome?

**Would you like me to create the "Applicant Cards" text for you to print out, or does this outline work for your planning?**

## **Classroom Debrief: Tying it to 2026 Regulations**


1. **The "High Risk" Label:** Under the **EU AI Act**, this admissions model is "High Risk." If your school used it, you would have to prove you checked for the bias that just occurred in the classroom.
2. **Individual vs. Group Fairness:** If we admit **S1** over **C4**, we are being "Group Fair" (correcting for systemic bias) but **C4** might complain that it's "Individually Unfair" because their SAT score was higher. There is no easy math to solve this - it’s a policy decision.
3. **The Human in the Loop:** In 2026, we don't let the AI make the final "Deny." We use the AI to flag potential, but a human must look at the "Context" column (the "Why") to make the final call.
